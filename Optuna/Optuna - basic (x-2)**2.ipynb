```python
import optuna
```

```python
def objective(trial):
    x = trial.suggest_float("x", -10, 10)
    return (x - 2) ** 2
```

```python
def runStudy(sampler):
    study = optuna.create_study(sampler=sampler)
    study.optimize(objective, n_trials=100)
    best_params = study.best_params
    found_x = best_params["x"]
    print("{} Found x: {}, (x - 2)^2: {}".format(study.sampler.__class__.__name__, found_x, (found_x - 2) ** 2))
```

```python
runStudy(optuna.samplers.RandomSampler())
```

```python
runStudy(optuna.samplers.CmaEsSampler())
```

```python
runStudy(optuna.samplers.TPESampler())
```

```python
search_space = {"x": [-10, 0, 10]}
runStudy(optuna.samplers.GridSampler(search_space))
```
